```python
import sqlite3
import pandas as pd
from datetime import datetime
```

```python
conn = sqlite3.connect("/home/sumkin/sumkin-trading/data/cca_trades_paper.db")
df = pd.read_sql_query("SELECT * FROM cca_trades_paper", conn)
df["commission"] = (df["spot_price_enter"] + df["futures_price_enter"] + df["spot_price_exit"] + df["futures_price_exit"]) * df["vol"] * 0.003
df["profit"] = (df["spot_price_exit"] - df["spot_price_enter"] + df["futures_price_enter"] - df["futures_price_exit"]) * df["vol"]
df["enter_dt"] = pd.to_datetime(df["enter_dt"], format="%Y-%m-%d %H:%M:%S")
df["exit_dt"] = pd.to_datetime(df["exit_dt"], format="%Y-%m-%d %H:%M:%S")
df["exit-enter"] = df["exit_dt"] - df["enter_dt"]
df["current_length_trade"] = datetime.now() - df["enter_dt"]
```

```python
df[df["active"] == 1].sort_values(by="current_length_trade", ascending=False).head(520)
```
